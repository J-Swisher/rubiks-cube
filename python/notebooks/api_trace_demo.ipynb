# Single Cubie API Trace

This notebook calls the tiny Python API the same way a frontend eventually will: start with one orientation state, apply named transforms in order, and get every intermediate state for animation.

```python
from cube_kernel import animation_payload, identity, named_transforms, trace_sequence

start = identity()
moves = ["x+", "z+", "y+"]
```

The math-layer API returns NumPy matrices.

```python
transforms = named_transforms(moves)
states = trace_sequence(start, transforms)

for index, state in enumerate(states):
    print(f"state {index}")
    print(state)
    print()
```

The frontend-facing API returns JSON-ready lists.

```python
payload = animation_payload(start, moves)
payload
```

The frontend can animate from `states[i]` to `states[i + 1]` for each transform.

```python
for move, before, after in zip(payload["transforms"], payload["states"], payload["states"][1:]):
    print(move)
    print("before:", before)
    print("after: ", after)
    print()
```
